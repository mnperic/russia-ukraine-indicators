```python
# Import dependencies and setup
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import matplotlib.dates as matplotlibdates
import datetime
from matplotlib.dates import date2num
```

```python
# Read .csv files and store as DataFrames

inflation_mom_path = "Resources/inflation_month_over_month.csv"
inflation_mom_df = pd.read_csv(inflation_mom_path)

inflation_monthly_path = "Resources/inflation_monthly.csv"
inflation_monthly_df = pd.read_csv(inflation_monthly_path)

inflation_yoy_path = "Resources/inflation_year_over_year.csv"
inflation_yoy_df = pd.read_csv(inflation_yoy_path)

gas_price_path = "Resources/gas_dutch.csv"
gas_df = pd.read_csv(gas_price_path)

oil_espo_path = "Resources/oil_espo.csv"
oil_espo_df = pd.read_csv(oil_espo_path)

oil_sokol_path = "Resources/oil_sokol.csv"
oil_sokol_df = pd.read_csv(oil_sokol_path)
```

```python
# Convert all 'date' columns to datetime

inflation_mom_df['date'] = pd.to_datetime(inflation_mom_df['date'], format='%d/%m/%y')
inflation_monthly_df['date'] = pd.to_datetime(inflation_monthly_df['date'], format='%d/%m/%y')
inflation_yoy_df['date'] = pd.to_datetime(inflation_yoy_df['date'], format='%d/%m/%y')
gas_df['date'] = pd.to_datetime(gas_df['date'], format='%d/%m/%y')
oil_espo_df['date'] = pd.to_datetime(oil_espo_df['date'], format='%d/%m/%y')
oil_sokol_df['date'] = pd.to_datetime(oil_sokol_df['date'], format='%d/%m/%y')
```

```python
# Plotting  inflation_mom

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.bar(inflation_mom_df['date'],inflation_mom_df['rate'], width=20, color='blue', alpha=0.5)
plt.plot(inflation_mom_df['date'], inflation_mom_df['rate'], color='orange', linewidth=3)

# plt.plot(inflation_mom_df['date'],inflation_mom_df['rate'])
# for a,b in zip(inflation_mom_df['date'], inflation_mom_df['rate']):
#     plt.text(a, b, str(b))

# Labels
plt.title("Inflation (month over month), Russia, Jan 2021  Jan 2022", fontdict = font_1)
plt.xlabel("Month", fontdict = font_2)
plt.ylabel("Growth Rate (%)", fontdict = font_2)

# Aesthetics
matplotlib.pyplot.grid(visible=True, which='both', axis='both', color='blue', alpha=0.1, linestyle='--', linewidth=1)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Save and show
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/russia-energy-inflation/Images/inflation_mom.png", dpi=300)
plt.show()
```

```python
# Plotting  gas_dutch

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.plot(gas_df['date'], gas_df['price_usd_mw'], color='blue', linewidth=3)

# Labels
plt.title("Dutch TTF Natural Gas Price, 2 Jan 2022  4 Mar 2022", fontdict = font_1)
plt.xlabel("Date", fontdict = font_2)
plt.ylabel("USD per mega-watt", fontdict = font_2)

# Highlights
plt.axvspan(date2num(datetime.datetime(2022,2,24)), date2num(datetime.datetime(2022,3,2)), 
           label="Invasion of Ukraine", color="red", alpha=0.05)
plt.legend()

# Aesthetics
matplotlib.pyplot.grid(visible=True, which='both', axis='both', color='blue', alpha=0.1, linestyle='--', linewidth=1)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Save and show
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/russia-energy-inflation/Images/gas_dutch.png", dpi=300)
plt.show()
```
